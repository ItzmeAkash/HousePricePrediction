### **Train Test Split**

```python
# Independent and dependent features

X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
##Train Test Split
from sklearn.model_selection import train_test_split\

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
##Standardize the dataset

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test )
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

###**Model Training**

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
##Print the coefficients and intercept

print(regression.coef_)
print(regression.intercept_)
```

```python
# On which parameters the model has been trained
regression.get_params()
```

```python
# Prediction with Test data

reg_predict=regression.predict(X_test)
reg_predict
```

### **Assumption**

```python
# plot a scatter plot for the prediction
plt.scatter(y_test,reg_predict)
```

```python
residuals=y_test-reg_predict
```

```python
residuals
```

```python
# plot this residual

sns.displot(residuals,kind='kde')
```

```python
# Scatter plot with respect to prediction and residuals
# Unifrom distribution
plt.scatter(reg_predict,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_predict))
print(mean_squared_error(y_test,reg_predict))
print(np.sqrt(mean_squared_error(y_test,reg_predict)))
```

### **R square and adjusted R square**

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_predict)
print(score)
```

```python
# display adjusted R squre

# adj_r2=1-[(1-r2)*(n-1)/n-k-1]


adjusted_R_squre=1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)

# adjusted r squre is less than r square this propertites is need to happend so it is current

print(adjusted_R_squre)
```
